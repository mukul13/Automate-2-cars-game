```python
"""
 
All coordinates assume a screen resolution of 1366*768, and Chrome 
maximized with the Bookmarks Toolbar disabled.
x_pad = 450
y_pad = 76
Play area =  x_pad+1, y_pad+1, 882, 723
"""

import ImageGrab
import os
import time
import win32api, win32con
import ImageOps

x_pad = 450
y_pad = 76
#Play area =  x_pad, y_pad, 882, 723

def screenGrab():
    #time.sleep(5)
    box = (x_pad,y_pad,x_pad+432,y_pad+647)
    im = ImageGrab.grab(box)
# im.save(os.getcwd() + '\\full_snap__' + str(int(time.time())) +
#'.png', 'PNG')
    return im


def leftClick():
    win32api.mouse_event(win32con.MOUSEEVENTF_LEFTDOWN,0,0)
    #time.sleep(.05)
    win32api.mouse_event(win32con.MOUSEEVENTF_LEFTUP,0,0)
    print "Click."          #completely optional. But nice for debugging purposes.



def mousePos(cord):
    win32api.SetCursorPos((x_pad + cord[0], y_pad + cord[1]))


def get_cords():
    x,y = win32api.GetCursorPos()
    x = x - x_pad
    y = y - y_pad
    print x,y
    
#get_cords()
# PLAY button 283 450
# right car 276 584
# left car 118 596
# game over 234, 84, 78
#54,160

def left_side_click():
    mousePos((118,455))
    leftClick()

def right_side_click():
    mousePos((276,455))
    leftClick()


left = 0
right = 1
# ll 89 386
# lr 175 386
# rl  246 386
# rr 342 386

def check_game_over(im):
    if(im.getpixel((90,605))!=(156, 162, 155)):
        return 1
    else:
        return 0
    

def check_left_obstacle(im):
    global left
    global right
    for j in range(1):
        for i in range(200):    
    # left obstacle and car in left
            if((im.getpixel((j+91,i+250))==(42,35,0)) and (left == 0)):
                print("l. left obstacle cleared!"+str(left))
                left=(left+1)%2
                left_side_click()
                #time.sleep(0.01)
                break;
    # right obstacle and car in right    
            if((im.getpixel((j+179,i+250))==(42,35,0)) and (left==1)):
                print("r. left obstacle cleared!"+str(left))
                left=(left+1)%2
                left_side_click()
                #time.sleep(0.01)
                break;
        

def check_right_obstacle(im):
    global left
    global right
    # left obstacle and car in left
    for j in range(1):
        for i in range(200):
            if((im.getpixel((246+j,i+250))==(0,13,42)) and (right == 0)):
                print("l. right obstacle cleared!"+str(right))
                right=(right+1)%2
                right_side_click()
                #time.sleep(0.01)
                break;
    # right obstacle and car in right    
            if((im.getpixel((345+j,i+250))==(0,13,42)) and (right == 1)):
                print("r. right obstacle cleared!"+str(right))
                right=(right+1)%2
                right_side_click()
                #time.sleep(0.01)
                break;
        
def check_left_star(im):
    global left
    global right
    # left star and car in right
    for j in range(1):
        for i in range(200):
            if((im.getpixel((89+j,250+i))==(255,210,63) or im.getpixel((89+j,230+i))== (223, 173, 52) ) and (left == 1)):
                print("Got left star!"+str(left))
                left=(left+1)%2
                left_side_click()
                break
    # right star and car in left
            if((im.getpixel((175+j,250+i))==(255,210,63) or im.getpixel((89+j,230+i))== (223, 173, 52)) and (left==0)):
                print("Got left star!"+str(left))
                left=(left+1)%2
                left_side_click()
                break

# blue star 69 181 248 
# left obstacle 42 35 0
# yellow star 255 210 63
# right obstacle 0 13 42        
        
def check_right_star(im):
    global left
    global right
    for j in range(1):
        for i in range(200):
    # left star and car in right    
            if((im.getpixel((246+j,250+i))==(69,181,248) or im.getpixel((246+j,230+i))==(56, 149, 203)) and (right == 1)):
                print("Got right star!"+str(right))
                right=(right+1)%2
                right_side_click()
    # right star and car in left    
            if((im.getpixel((345+j,250+i)) ==(69,181,248) or im.getpixel((345+j,230+i))==(56, 149, 203)) and (right==0)):
                print("Got right star!"+str(right))
                right=(right+1)%2
                right_side_click()



def play_game():
    mousePos((283,450))
    leftClick()
    time.sleep(2)
    
    while 1:
        im = screenGrab()
        #print(str(int(time.time())))
        #if(check_game_over(im)==1):
        #    print("Game Over")
         #   break;
        
        check_left_obstacle(im)
        check_right_obstacle(im)
        check_left_star(im)
        check_right_star(im)
        #time.sleep(0.01)
    
    
```

```python
play_game()
```
